```python
r = 0.01       
sigma = 0.15   
T = 10          
t = 0          
S0 = 100 
n = 100
guarantee = 100
```

```python
def get_optionPrice(r, sigma, T, guarantee, n):

    # Formulas from slides
    dt = T/n
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u

    R_dt = np.exp(r*dt) # growth factor at each time step
    p = (R_dt - d) / (u - d) # Risk-neutral probability
    q = 1 - p


    # Stock price calculations
    stockValues = np.zeros((n+1, n+1))
    stockValues[0,0] = 100 # Initial value
    
    for i in range(1,n+1):
        stockValues[i,0] = u*stockValues[i-1,0]
        for j in range(1,i+1):
           stockValues[i,j] = d*stockValues[i-1,j-1]



    # contract price calculations
    contractPrice = np.zeros((n+1, n+1))

    for i in range(n+1):
         contractPrice[n,i] = np.maximum(guarantee,stockValues[n,i]) 
         
    for i in range(n - 1, -1, -1):
        for j in range(i+1):
            contractPrice[i,j] = (1/R_dt)*(p*contractPrice[i+1,j] + q*contractPrice[i+1,j+1])

    initialPrice = contractPrice[0,0]
       
    return initialPrice
```

```python
step_counts = range(1, 100)
prices = [get_optionPrice(r, sigma, T, guarantee, n) for n in step_counts]

print("Price of the Contract: ", get_optionPrice(r, sigma, T, guarantee, n)) 

analyticalPrice = 113.482

plt.figure(figsize=(10, 6))
plt.plot(step_counts, prices, label='Binomial Tree')
plt.axhline(y=analyticalPrice, color='red', linestyle='--', label='Analytical Price')
plt.xlabel('Number of Steps')
plt.ylabel('Price of the Contract')
plt.title('Price of the Contract with increasing number of steps')
plt.legend()
plt.show()
```

# Delta-hedge

```python
def binomial_tree_slow(K, T, S0, r, N, sigma):
    # precompute constants
    dt = T / N
    R = np.exp(r * dt)
    disc = 1 / R
    u = R * np.exp(sigma * np.sqrt(dt))   # up factor
    d = R * np.exp(-sigma * np.sqrt(dt))  # down factor
    q = (R - d) / (u - d)

    # initialise asset prices at maturity - Time step N
    S = np.zeros(N + 1)
    S[0] = S0 * d**N
    for j in range(1, N + 1):
        S[j] = S[j - 1] * u / d

    # initialise option values at maturity
    C = np.zeros(N + 1)
    for j in range(0, N + 1):
        C[j] = max(K, S[j])  # max(0, S[j] - K)

    # step backwards through tree
    # D = np.zeros(N + 1)
    for i in np.arange(N, 0, -1):
        for j in range(0, i):
            # D[j] = (C[j + 1] - C[j]) / (S[j + 1] - S[j])
            C[j] = disc * (q * C[j + 1] + (1 - q) * C[j])

    # delta hedge
    ftu = C[1]
    ftd = ((C[0] / disc) - (q * ftu)) / (1 - q)

    Stu = S0 * u
    Std = S0 * d

    DH = (ftu - ftd) / (Stu - Std)

    return C[0], DH
```

```python
r = 0.01       
sigma = 0.15   
T = 10          
t = 0          
S0 = 100 
n = 100
guarantee = 100
C, DH = binomial_tree_slow(100, T, S0, r,n, sigma)
print(C, DH)
```

```python
def plot_binomial_tree(stock_prices, deltas, n):
    plt.figure(figsize=(12, 8))
    
    # Plot nodes and lines
    for i in range(n + 1):
        for j in range(i + 1):
            # Plot nodes
            plt.plot(i, stock_prices[i][j], 'bo')
            
            # Plot connecting lines
            if i < n:
                # Plot lines to the next time step's nodes
                plt.plot([i, i + 1], [stock_prices[i][j], stock_prices[i + 1][j]], 'b-')
                plt.plot([i, i + 1], [stock_prices[i][j], stock_prices[i + 1][j + 1]], 'b-')
            
            # Annotate delta values
            delta_val = deltas[i][j]
            plt.annotate(f'={delta_val:.2f}', (i, stock_prices[i][j]),
                         textcoords="offset points", xytext=(0,10), ha='center')

    plt.title("Binomial Tree for Stock Prices and Delta Values")
    plt.xlabel("Time Step")
    plt.ylabel("Stock Price")
    plt.show()

# Initialize parameters
S0 = 100  # Initial stock price
u = 1.2   # Up factor
d = 0.8   # Down factor
n = 4     # Number of time steps

# Calculate stock prices
stock_prices = [[0 for j in range(i + 1)] for i in range(n + 1)]
stock_prices[0][0] = S0

for i in range(1, n + 1):
    for j in range(i + 1):
        # Calculate stock price at each node
        stock_prices[i][j] = S0 * (u ** (i - j)) * (d ** j)

# Sample delta values (you would typically calculate these based on option parameters)
deltas = [[0 for j in range(i + 1)] for i in range(n + 1)]
deltas[0][0] = 0.67
deltas[1][0] = 0.78
deltas[1][1] = 0.52
deltas[2][0] = 0.87
deltas[2][1] = 0.66
deltas[2][2] = 0.36
deltas[3][0] = 0.93
deltas[3][1] = 0.77
deltas[3][2] = 0.50
deltas[3][3] = 0.23
deltas[4][0] = 0.98
deltas[4][1] = 0.88
deltas[4][2] = 0.64
deltas[4][3] = 0.32
deltas[4][4] = 0.10

# Plot the tree
plot_binomial_tree(stock_prices, deltas, n)
```

```python
import numpy as np
from scipy.stats import norm

# Parameters
S0 = 100  # Initial stock price
K = 100   # Strike price
r = 0.01  # Risk-free rate
sigma = 0.15  # Volatility
T = 10    # Time to maturity
G = 100   # Guaranteed amount

def closed_form_delta(S, t):
    """Calculate closed-form delta at time t for stock price S"""
    tau = T - t
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*tau)/(sigma*np.sqrt(tau))
    return norm.cdf(d1)

def binomial_tree_delta(steps):
    """Calculate delta using binomial tree with given number of steps"""
    dt = T/steps
    u = np.exp(sigma*np.sqrt(dt))
    d = 1/u
    p = (np.exp(r*dt) - d)/(u - d)
    
    # Initialize stock price tree
    stock = np.zeros((steps+1, steps+1))
    stock[0,0] = S0
    
    # Forward pass - fill in stock price tree
    for j in range(1, steps+1):
        for i in range(j+1):
            stock[i,j] = S0 * (u**(j-i)) * (d**i)
    
    # Initialize option value tree
    option = np.zeros((steps+1, steps+1))
    
    # Terminal payoff
    for i in range(steps+1):
        option[i,steps] = max(stock[i,steps], G)
    
    # Backward induction
    for j in range(steps-1, -1, -1):
        for i in range(j+1):
            option[i,j] = np.exp(-r*dt)*(p*option[i,j+1] + (1-p)*option[i+1,j+1])
    
    # Calculate delta at t=0
    delta = (option[0,1] - option[1,1])/(stock[0,1] - stock[1,1])
    
    return delta

# Calculate closed-form delta at t=0
closed_form_delta_0 = closed_form_delta(S0, 0)
print(f"Closed-form delta at t=0: {closed_form_delta_0:.4f}")

# Calculate and compare delta for different numbers of steps
steps_list = [10, 50, 100, 500, 1000]
print("\nBinomial tree deltas:")
for steps in steps_list:
    delta = binomial_tree_delta(steps)
    print(f"Steps: {steps:4d}, Delta: {delta:.4f}, Difference from closed-form: {abs(delta-closed_form_delta_0):.6f}")

# Plot convergence of binomial delta to closed-form delta
import matplotlib.pyplot as plt

steps_for_plot = range(10, 201, 5)
deltas = [binomial_tree_delta(n) for n in steps_for_plot]

plt.figure(figsize=(10, 6))
plt.plot(steps_for_plot, deltas, label='Binomial Tree Delta')
plt.axhline(y=closed_form_delta_0, color='r', linestyle='--', label='Closed-form Delta')
plt.xlabel('Number of Steps')
plt.ylabel('Delta')
plt.title('Convergence of Binomial Tree Delta to Closed-form Delta')
plt.legend()
plt.grid(True)
plt.show()

# Calculate delta for different stock prices at t=0
stock_prices = np.linspace(80, 120, 41)
closed_form_deltas = [closed_form_delta(S, 0) for S in stock_prices]

plt.figure(figsize=(10, 6))
plt.plot(stock_prices, closed_form_deltas)
plt.xlabel('Stock Price')
plt.ylabel('Delta')
plt.title('Delta vs Stock Price at t=0')
plt.grid(True)
plt.show()
```
